## Popcorn Hack #1: Transform to Compound Assignment Operators

**Task:** Transform beginner code into more advanced code using compound assignment operators.

### Original Code (Beginner Style):

```java
public class BeginnerCode {
    public static void main(String[] args) {
        System.out.println("=== Original Beginner Code ===");
        System.out.println();
        
        int playerScore = 1000;
        int playerHealth = 100;
        int enemiesDefeated = 0;
        
        System.out.println("Starting values:");
        System.out.println("Score: " + playerScore);
        System.out.println("Health: " + playerHealth);
        System.out.println("Enemies Defeated: " + enemiesDefeated);
        System.out.println();

        // Player defeats an enemy worth 250 points
        playerScore = playerScore + 250;
        System.out.println("After defeating enemy: Score = " + playerScore);

        // Player takes 15 damage
        playerHealth = playerHealth - 15;
        System.out.println("After taking damage: Health = " + playerHealth);

        // Enemy count goes up
        enemiesDefeated = enemiesDefeated + 1;
        System.out.println("Enemies defeated: " + enemiesDefeated);

        // Boss battle: double the current score!
        playerScore = playerScore * 2;
        System.out.println("After boss battle: Score = " + playerScore);

        // Healing potion restores health to 80% of current
        playerHealth = playerHealth * 4;
        playerHealth = playerHealth / 5;  // 4/5 = 0.8
        System.out.println("After healing potion: Health = " + playerHealth);
    }
}

BeginnerCode.main(null);
```

### Transformed Code (Advanced Style with Compound Operators):

```java
public class AdvancedCode {
    public static void main(String[] args) {
        System.out.println("=== Advanced Code with Compound Operators ===");
        System.out.println();
        
        int playerScore = 1000;
        int playerHealth = 100;
        int enemiesDefeated = 0;
        
        System.out.println("Starting values:");
        System.out.println("Score: " + playerScore);
        System.out.println("Health: " + playerHealth);
        System.out.println("Enemies Defeated: " + enemiesDefeated);
        System.out.println();

        // Player defeats an enemy worth 250 points
        playerScore += 250;  // Changed from: playerScore = playerScore + 250
        System.out.println("After defeating enemy: Score = " + playerScore);

        // Player takes 15 damage
        playerHealth -= 15;  // Changed from: playerHealth = playerHealth - 15
        System.out.println("After taking damage: Health = " + playerHealth);

        // Enemy count goes up
        enemiesDefeated++;  // Changed from: enemiesDefeated = enemiesDefeated + 1
        System.out.println("Enemies defeated: " + enemiesDefeated);

        // Boss battle: double the current score!
        playerScore *= 2;  // Changed from: playerScore = playerScore * 2
        System.out.println("After boss battle: Score = " + playerScore);

        // Healing potion restores health to 80% of current
        playerHealth *= 4;  // Changed from: playerHealth = playerHealth * 4
        playerHealth /= 5;  // Changed from: playerHealth = playerHealth / 5
        System.out.println("After healing potion: Health = " + playerHealth);
        
        System.out.println();
        System.out.println("=== Transformations Made ===");
        System.out.println("‚úì playerScore + 250     ‚Üí  playerScore += 250");
        System.out.println("‚úì playerHealth - 15     ‚Üí  playerHealth -= 15");
        System.out.println("‚úì enemiesDefeated + 1   ‚Üí  enemiesDefeated++");
        System.out.println("‚úì playerScore * 2       ‚Üí  playerScore *= 2");
        System.out.println("‚úì playerHealth * 4      ‚Üí  playerHealth *= 4");
        System.out.println("‚úì playerHealth / 5      ‚Üí  playerHealth /= 5");
    }
}

AdvancedCode.main(null);
```

### Benefits of Compound Assignment Operators:

1. **More Concise** - Less typing and cleaner code
2. **Easier to Read** - Intent is immediately clear
3. **Less Error-Prone** - Don't have to repeat variable names
4. **Professional Style** - Used in industry-standard code
5. **Slightly More Efficient** - Variable is evaluated once instead of twice

## Popcorn Hack #2: Score Manipulation Program

**Task:** Write a program where `score` starts at 100 and use at least three different compound assignment operators.

**Goals:**
- Deduct points for a wrong answer (-=)
- Double the score with a power-up (*=)
- Find remainder after dividing by 7 (%=)

```java
public class ScoreManipulation {
    public static void main(String[] args) {
        System.out.println("=== Quiz Game Score Manipulation ===");
        System.out.println();
        
        int score = 100;
        System.out.println("Starting score: " + score);
        System.out.println("=".repeat(40));
        System.out.println();
        
        // 1. Add points for a correct answer (+=)
        System.out.println("Question 1: What is 5 + 7?");
        System.out.println("Answer: Correct! +25 points");
        score += 25;
        System.out.println("Score after Q1: " + score);
        System.out.println();
        
        // 2. Deduct points for a wrong answer (-=)
        System.out.println("Question 2: What is the capital of Mars?");
        System.out.println("Answer: Wrong! -30 points");
        score -= 30;
        System.out.println("Score after Q2: " + score);
        System.out.println();
        
        // 3. Double the score with a power-up (*=)
        System.out.println("üéÅ BONUS POWER-UP ACTIVATED!");
        System.out.println("Your score doubles!");
        score *= 2;
        System.out.println("Score after power-up: " + score);
        System.out.println();
        
        // 4. Bonus: Add combo points (+=)
        System.out.println("‚ö° Combo bonus: +50 points");
        score += 50;
        System.out.println("Score after combo: " + score);
        System.out.println();
        
        // 5. Calculate difficulty adjustment (/=)
        System.out.println("Adjusting for difficulty level...");
        System.out.println("Dividing by 3 for final adjusted score");
        score /= 3;
        System.out.println("Adjusted score: " + score);
        System.out.println();
        
        // 6. Find remainder after dividing by 7 (%=)
        System.out.println("Finding position in weekly leaderboard...");
        System.out.println("Calculating: score % 7");
        score %= 7;
        System.out.println("Leaderboard position (0-6): " + score);
        System.out.println();
        
        System.out.println("=".repeat(40));
        System.out.println("Final position: " + score);
        System.out.println();
        
        // Summary of operators used
        System.out.println("Compound Operators Used:");
        System.out.println("‚úì +=  (Addition assignment)");
        System.out.println("‚úì -=  (Subtraction assignment)");
        System.out.println("‚úì *=  (Multiplication assignment)");
        System.out.println("‚úì /=  (Division assignment)");
        System.out.println("‚úì %=  (Modulus assignment)");
    }
}

ScoreManipulation.main(null);
```

## Homework Assignment: Fitness Progress Tracker (Option C)

I chose **Option C: Fitness Progress Tracker** to model a fitness journey with daily health metrics.

### Variables Tracked:
- `totalCalories` - Net calories (consumed - burned)
- `workoutDays` - Number of days exercised
- `stepCount` - Daily step count
- `waterIntake` - Glasses of water consumed
- `weekProgress` - Days into current fitness week

```java
public class FitnessProgressTracker {
    public static void main(String[] args) {
        System.out.println("=".repeat(50));
        System.out.println("        FITNESS PROGRESS TRACKER");
        System.out.println("=".repeat(50));
        System.out.println();
        
        // Initialize tracking variables
        int totalCalories = 2000;    // Starting daily calorie target
        int workoutDays = 0;         // Days exercised this week
        int stepCount = 5000;        // Steps taken today
        int waterIntake = 4;         // Glasses of water consumed
        int weekProgress = 1;        // Current day of fitness challenge
        
        System.out.println("STARTING METRICS");
        System.out.println("-".repeat(50));
        System.out.println("Total Calories: " + totalCalories + " kcal");
        System.out.println("Workout Days: " + workoutDays);
        System.out.println("Step Count: " + stepCount);
        System.out.println("Water Intake: " + waterIntake + " glasses");
        System.out.println("Week Progress: Day " + weekProgress);
        System.out.println();
        System.out.println("=".repeat(50));
        System.out.println();
        
        // Morning: Breakfast
        System.out.println("MORNING ROUTINE");
        System.out.println("-".repeat(50));
        System.out.println("Ate a healthy breakfast (oatmeal, banana, eggs)");
        totalCalories += 450;  // Add breakfast calories
        System.out.println("Calories consumed: +450 kcal");
        System.out.println("Total Calories: " + totalCalories + " kcal");
        System.out.println();
        
        System.out.println("Drank morning water");
        waterIntake++;  // Increment water count
        System.out.println("Water Intake: " + waterIntake + " glasses");
        System.out.println();
        
        // Midday: Workout Session
        System.out.println("WORKOUT SESSION");
        System.out.println("-".repeat(50));
        System.out.println("Completed 45-minute cardio workout!");
        totalCalories -= 400;  // Burn calories from cardio
        System.out.println("Calories burned: -400 kcal");
        System.out.println("Net Calories: " + totalCalories + " kcal");
        System.out.println();
        
        workoutDays++;  // Increment workout days
        System.out.println("Workout days this week: " + workoutDays);
        System.out.println();
        
        // Afternoon: Extra Active Day
        System.out.println("üö∂ AFTERNOON ACTIVITY");
        System.out.println("-".repeat(50));
        System.out.println("Went for a long walk - extra active today!");
        stepCount *= 2;  // Double steps from being very active
        System.out.println("Step count doubled: " + stepCount + " steps");
        System.out.println();
        
        System.out.println("Lunch time - healthy salad with grilled chicken");
        totalCalories += 550;
        System.out.println("Calories consumed: +550 kcal");
        System.out.println("Total Calories: " + totalCalories + " kcal");
        System.out.println();
        
        waterIntake += 2;  // Drank 2 more glasses
        System.out.println("Stayed hydrated: +2 glasses");
        System.out.println("Water Intake: " + waterIntake + " glasses");
        System.out.println();
        
        // Evening: Dinner and Rest
        System.out.println("EVENING ROUTINE");
        System.out.println("-".repeat(50));
        System.out.println("Dinner - balanced meal with protein and vegetables");
        totalCalories += 600;
        System.out.println("Calories consumed: +600 kcal");
        System.out.println("Total Calories: " + totalCalories + " kcal");
        System.out.println();
        
        System.out.println("Light evening walk");
        totalCalories -= 150;  // Burn some evening calories
        System.out.println("Calories burned: -150 kcal");
        System.out.println("Net Calories: " + totalCalories + " kcal");
        System.out.println();
        
        waterIntake++;  // Final glass before bed
        System.out.println("Final glass of water before bed");
        System.out.println("Water Intake: " + waterIntake + " glasses");
        System.out.println();
        
        // Weekly Calculations
        System.out.println("WEEKLY STATISTICS");
        System.out.println("-".repeat(50));
        
        // Calculate average daily steps for the week
        int weeklySteps = stepCount;
        System.out.println("Calculating average daily steps over 7 days...");
        weeklySteps /= 7;  // Average steps per day
        System.out.println("Average daily steps: " + weeklySteps + " steps");
        System.out.println();
        
        // Track progress through the week (day 1-7)
        System.out.println("Checking week progress position...");
        weekProgress %= 7;  // Keep day within 0-6 range for week tracking
        System.out.println("Day of week (0=Sunday): " + weekProgress);
        System.out.println();
        
        // Final Summary
        System.out.println("=".repeat(50));
        System.out.println("        DAILY SUMMARY");
        System.out.println("=".repeat(50));
        System.out.println();
        System.out.println("‚úì Final Net Calories: " + totalCalories + " kcal");
        System.out.println("‚úì Total Steps Today: " + stepCount + " steps");
        System.out.println("‚úì Water Consumed: " + waterIntake + " glasses");
        System.out.println("‚úì Workout Days This Week: " + workoutDays);
        System.out.println("‚úì Average Daily Steps: " + weeklySteps + " steps");
        System.out.println();
        
        // Motivational feedback
        System.out.println("PROGRESS FEEDBACK");
        System.out.println("-".repeat(50));
        
        if (totalCalories >= 2000 && totalCalories <= 2500) {
            System.out.println("Perfect calorie balance! You're on track!");
        } else if (totalCalories < 2000) {
            System.out.println("‚ö†Ô∏è  Consider eating more to meet your energy needs.");
        } else {
            System.out.println("‚ö†Ô∏è  Slightly over calorie goal, but that's okay!");
        }
        
        if (stepCount >= 10000) {
            System.out.println("Excellent! You hit the 10,000 step goal!");
        } else {
            System.out.println("Keep moving! Try to reach 10,000 steps tomorrow.");
        }
        
        if (waterIntake >= 8) {
            System.out.println("Great hydration! You're well hydrated.");
        } else {
            System.out.println("Drink more water to stay hydrated!");
        }
        
        System.out.println();
        System.out.println("=".repeat(50));
        System.out.println("     Keep up the great work!");
        System.out.println("=".repeat(50));
        System.out.println();
        
        // Compound operators summary
        System.out.println("COMPOUND OPERATORS USED:");
        System.out.println("-".repeat(50));
        System.out.println("‚úì +=  Add calories from meals");
        System.out.println("‚úì -=  Subtract calories from exercise");
        System.out.println("‚úì *=  Double step count on active days");
        System.out.println("‚úì /=  Calculate average daily steps");
        System.out.println("‚úì %=  Track weekly progress position");
        System.out.println("‚úì ++  Increment water intake and workout days");
        System.out.println();
    }
}

FitnessProgressTracker.main(null);
```

## Homework Summary

### Compound Assignment Operators Used:

| Operator | Usage in Program | Purpose |
|----------|------------------|----------|
| `+=` | `totalCalories += 450` | Add calories from meals |
| `-=` | `totalCalories -= 400` | Subtract calories burned from exercise |
| `*=` | `stepCount *= 2` | Double step count on extra active days |
| `/=` | `weeklySteps /= 7` | Calculate average daily steps over a week |
| `%=` | `weekProgress %= 7` | Keep week progress within 0-6 range |
| `++` | `workoutDays++` | Increment workout day counter |
| `++` | `waterIntake++` | Increment water glass counter |

### Real-World Application:

This fitness tracker demonstrates how compound assignment operators make code:
- **More readable** - Intent is immediately clear
- **More maintainable** - Less chance of typos
- **More professional** - Follows industry best practices
- **More efficient** - Variables evaluated once instead of twice

### Key Features:

**Meaningful variable names** - Clear purpose for each variable
**Print statements after each operation** - Shows progression
**Detailed comments** - Explains what each operation represents
**Complete simulation** - Models a full day of fitness tracking
**Motivational feedback** - Provides encouraging messages based on progress
**Multiple compound operators** - Uses all required operators (+=, -=, *=, /=, %=, ++)

### Important Concepts Demonstrated:

1. **Integer Division** - `weeklySteps /= 7` truncates to integer result
2. **Modulus for Boundaries** - `%=` keeps values within range (0-6 for week days)
3. **Increment Operators** - `++` is shorthand for adding 1
4. **Accumulation** - Using `+=` to build running totals
5. **Reduction** - Using `-=` to subtract from totals
