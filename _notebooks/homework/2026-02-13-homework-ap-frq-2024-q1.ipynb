## Challenge #1

```java
public class Feeder {

    private int currentFood;

    public Feeder(int initialFood) {
        this.currentFood = initialFood;
    }

    public void simulateOneDay(int numBirds) {
        if (Math.random() < 0.05) {
            currentFood = 0;
        } else {
            int quantity = (int) (Math.random() * 41) + 10;
            int total = numBirds * quantity;
            if (total > currentFood) {
                currentFood = 0;
            } else {
                currentFood -= total;
            }
        }
    }

    public int simulateManyDays(int numBirds, int numDays) {
        int activeDays = 0;
        for (int i = 0; i < numDays; i++) {
            if (currentFood <= 0) break;
            activeDays++;
            simulateOneDay(numBirds);
        }
        return activeDays;
    }

    public int getCurrentFood() {
        return currentFood;
    }
}

class Main {
    public static void main(String[] args) {
        Feeder f = new Feeder(500);
        System.out.println("Initial food: " + f.getCurrentFood());
        f.simulateOneDay(10);
        System.out.println("After 1 day: " + f.getCurrentFood());
        
        Feeder f2 = new Feeder(2400);
        int days = f2.simulateManyDays(10, 4);
        System.out.println("Days with food: " + days);
    }
}

Main.main(null);
```

## How I Solved it

I structured the `Feeder` class to maintain the `currentFood` state. For the daily simulation, I prioritized the 5% bear check using `Math.random()`. If that didn't trigger, I calculated the bird consumption by generating a random value in the 10-50 range and multiplying by the number of birds. In the multi-day simulation, I used a loop that checks the food level at the start of each day to decide whether to increment the count and run the simulation.

## Area of Struggle

My main difficulty was ensuring the multi-day loop didn't count a day when the feeder was starting at zero. I initially put the `simulateOneDay` call before the check, which gave me an extra day in the results. I had to rethink the order so that the "finding food" condition was checked at the very beginning of the loop iteration.
