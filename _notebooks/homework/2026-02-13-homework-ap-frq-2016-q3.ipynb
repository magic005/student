## Challenge #1

```java
class Square {
    private boolean isBlack;
    private int num;
    
    public Square(boolean isBlack, int num) {
        this.isBlack = isBlack;
        this.num = num;
    }
    
    @Override
    public String toString() {
        return isBlack ? "[X]" : String.valueOf(num);
    }
}

public class Crossword {
    private Square[][] puzzle;

    private boolean toBeLabeled(int r, int c, boolean[][] blackSquares) {
        if (blackSquares[r][c]) return false;
        if (r == 0 || c == 0) return true;
        return blackSquares[r - 1][c] || blackSquares[r][c - 1];
    }

    public Crossword(boolean[][] blackSquares) {
        int rows = blackSquares.length;
        int cols = blackSquares[0].length;
        this.puzzle = new Square[rows][cols];
        int nextNum = 1;

        for (int r = 0; r < rows; r++) {
            for (int c = 0; c < cols; c++) {
                if (toBeLabeled(r, c, blackSquares)) {
                    puzzle[r][c] = new Square(false, nextNum++);
                } else {
                    puzzle[r][c] = new Square(blackSquares[r][c], 0);
                }
            }
        }
    }

    public void print() {
        for (Square[] row : puzzle) {
            for (Square s : row) System.out.print(s + "\t");
            System.out.println();
        }
    }
}

class Main {
    public static void main(String[] args) {
        // NOTE: added spaces in { { to avoid Liquid errors
        boolean[][] grid = { 
            {true, false, false}, 
            {false, false, true}, 
            {false, true, false} 
        };
        Crossword cw = new Crossword(grid);
        cw.print();
    }
}

Main.main(null);
```

## How I Solved it

I approached the crossword problem by first defining the `Square` object to handle the dual state of color and numbering. For the labeling logic in `toBeLabeled`, I simplified the conditions: if a square is black, it's never labeled. If it's on the top or left edge, it's always labeled. Otherwise, it just checks the neighbors. In the constructor, I used a row-major traversal and a simple incrementing counter which only advances when a label is assigned.

## Area of Struggle

The biggest challenge was handling the sequential numbering. I initially tried to calculate the number based on the row and column indices, but that doesn't work because black squares skip numbers. I switched to a dedicated `nextNum` variable that only increments when a square satisfies the `toBeLabeled` condition, which fixed the sequence.

## Final result

The output grid shows `[X]` for black squares and correctly ordered integers for the white squares. The numbering matches the row-major requirement perfectly.
