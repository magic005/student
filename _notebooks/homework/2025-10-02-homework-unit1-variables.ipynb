## Popcorn Hack #1: My Favorite Things

Creating variables to store information about my favorite things with the appropriate data types.

```java
public class MyFavoriteThings {
    public static void main(String[] args) {
        // My favorite things using appropriate data types
        
        // 1. Favorite food (text)
        String favoriteinstrument = "guitar";
        
        // 2. My age (whole number)
        int myAge = 16;
        
        // 3. Height in feet (decimal)
        double heightInFeet = 5.11;
        
        // 4. Do I like dogs? (True/False)
        boolean likesdogs = true;
        
        // 5. First letter of favorite color
        char colorFirstLetter = 'L';  // Green
        
        // 6. Birth year (never changes - use final)
        final int BIRTH_YEAR = 2009;
        
        // Display all the information
        System.out.println("=== My Favorite Things ===");
        System.out.println("Favorite Instrument: " + favoriteinstrument);
        System.out.println("Age: " + myAge + " years old");
        System.out.println("Height: " + heightInFeet + " feet");
        System.out.println("Do I like dogs? " + likesdogs);
        System.out.println("Favorite color starts with: " + colorFirstLetter);
        System.out.println("Birth Year: " + BIRTH_YEAR);
        
        // Extra: Calculate current year
        int currentYear = 2025;
        int calculatedAge = currentYear - BIRTH_YEAR;
        System.out.println("\nVerification: " + currentYear + " - " + BIRTH_YEAR + " = " + calculatedAge);
    }
}

MyFavoriteThings.main(null);
```

### Explanation of Data Type Choices:

- **String** for favorite food - stores text/words
- **int** for age - whole number (no decimals needed)
- **double** for height - allows decimal values (5.8 feet)
- **boolean** for liking pizza - true or false value
- **char** for first letter - single character in single quotes
- **final int** for birth year - constant value that never changes

## Popcorn Hack #2: Pick the Best Type

Identifying and using the correct data type for each piece of information.

```java
public class PickTheType {
    public static void main(String[] args) {
        System.out.println("=== Data Type Selection Practice ===");
        System.out.println();
        
        // 1. Number of siblings - int (counting whole people)
        int numberOfSiblings = 3;
        System.out.println("1. Number of siblings (int): " + numberOfSiblings);
        System.out.println("   Why int? Counting people, always a whole number\n");
        
        // 2. Your first name - String (text)
        String firstName = "Jordan";
        System.out.println("2. First name (String): " + firstName);
        System.out.println("   Why String? Names are text/words\n");
        
        // 3. Are you hungry? - boolean (yes/no question)
        boolean isHungry = false;
        System.out.println("3. Are you hungry? (boolean): " + isHungry);
        System.out.println("   Why boolean? Yes/No question, true or false\n");
        
        // 4. Your favorite letter - char (single character)
        char favoriteLetter = 'J';
        System.out.println("4. Favorite letter (char): " + favoriteLetter);
        System.out.println("   Why char? Single character in single quotes\n");
        
        // 5. Your height in inches - double (could have decimals)
        double heightInches = 68.5;
        System.out.println("5. Height in inches (double): " + heightInches);
        System.out.println("   Why double? Height can have decimal values\n");
        
        // 6. Number of days in a year - final int (constant, never changes)
        final int DAYS_IN_YEAR = 365;
        System.out.println("6. Days in a year (final int): " + DAYS_IN_YEAR);
        System.out.println("   Why final int? Constant value that never changes\n");
        
        // Summary
        System.out.println("=" .repeat(50));
        System.out.println("Key Takeaway: Choose the data type that best");
        System.out.println("matches what you're storing!");
    }
}

PickTheType.main(null);
```

## Homework Hack: Simple Grade Calculator

Creating a complete grade calculator program that stores student information and calculates their final grade.

```java
public class GradeCalculator {
    public static void main(String[] args) {
        // Student information (using final for constants)
        final String STUDENT_NAME = "Emily Chen";
        final String CLASS_NAME = "AP Computer Science A";
        
        // Three test scores (whole numbers 0-100)
        int test1Score = 88;
        int test2Score = 92;
        int test3Score = 85;
        
        // Calculate the average
        double average = (test1Score + test2Score + test3Score) / 3.0;
        
        // Determine letter grade
        String letterGrade;
        if (average >= 90) {
            letterGrade = "A";
        } else if (average >= 80) {
            letterGrade = "B";
        } else if (average >= 70) {
            letterGrade = "C";
        } else if (average >= 60) {
            letterGrade = "D";
        } else {
            letterGrade = "F";
        }
        
        // Display the grade report
        System.out.println("\n" + "=".repeat(50));
        System.out.println("           STUDENT GRADE REPORT");
        System.out.println("=".repeat(50));
        System.out.println("Student Name: " + STUDENT_NAME);
        System.out.println("Class: " + CLASS_NAME);
        System.out.println("-".repeat(50));
        System.out.println("Test Scores:");
        System.out.println("  Test 1: " + test1Score);
        System.out.println("  Test 2: " + test2Score);
        System.out.println("  Test 3: " + test3Score);
        System.out.println("-".repeat(50));
        System.out.printf("Average Score: %.2f%%%n", average);
        System.out.println("Letter Grade: " + letterGrade);
        System.out.println("=".repeat(50));
        
        // Performance feedback
        System.out.println("\nPerformance Feedback:");
        if (letterGrade.equals("A")) {
            System.out.println("Outstanding work! Keep up the excellent performance!");
        } else if (letterGrade.equals("B")) {
            System.out.println("Great job! You're doing very well in this class.");
        } else if (letterGrade.equals("C")) {
            System.out.println("Good effort. Consider reviewing material for improvement.");
        } else if (letterGrade.equals("D")) {
            System.out.println("You're passing, but extra study time would be beneficial.");
        } else {
            System.out.println("Please see your teacher for additional support.");
        }
    }
}

GradeCalculator.main(null);
```

### Enhanced Grade Calculator with User Input

Bonus version that allows entering different test scores:

```java
import java.util.Scanner;

public class InteractiveGradeCalculator {
    public static void main(String[] args) {
        // Constants that never change
        final String CLASS_NAME = "AP Computer Science A";
        final int MAX_SCORE = 100;
        final int MIN_SCORE = 0;
        
        Scanner scanner = new Scanner(System.in);
        
        System.out.println("\n" + "=".repeat(50));
        System.out.println("     INTERACTIVE GRADE CALCULATOR");
        System.out.println("=".repeat(50));
        
        try {
            // Get student name
            System.out.print("\nEnter student name: ");
            final String STUDENT_NAME = scanner.nextLine();
            
            // Get three test scores
            System.out.print("Enter Test 1 score (0-100): ");
            int test1 = scanner.nextInt();
            
            System.out.print("Enter Test 2 score (0-100): ");
            int test2 = scanner.nextInt();
            
            System.out.print("Enter Test 3 score (0-100): ");
            int test3 = scanner.nextInt();
            
            // Validate scores
            if (test1 < MIN_SCORE || test1 > MAX_SCORE ||
                test2 < MIN_SCORE || test2 > MAX_SCORE ||
                test3 < MIN_SCORE || test3 > MAX_SCORE) {
                System.out.println("\nError: All scores must be between 0 and 100!");
                return;
            }
            
            // Calculate average
            double average = (test1 + test2 + test3) / 3.0;
            
            // Determine letter grade
            String letterGrade;
            if (average >= 90) {
                letterGrade = "A";
            } else if (average >= 80) {
                letterGrade = "B";
            } else if (average >= 70) {
                letterGrade = "C";
            } else if (average >= 60) {
                letterGrade = "D";
            } else {
                letterGrade = "F";
            }
            
            // Display results
            System.out.println("\n" + "=".repeat(50));
            System.out.println("           GRADE REPORT");
            System.out.println("=".repeat(50));
            System.out.println("Student: " + STUDENT_NAME);
            System.out.println("Class: " + CLASS_NAME);
            System.out.println("-".repeat(50));
            System.out.println("Test Scores: " + test1 + ", " + test2 + ", " + test3);
            System.out.printf("Average: %.2f%%%n", average);
            System.out.println("Final Grade: " + letterGrade);
            System.out.println("=".repeat(50));
            
            // Calculate points to next grade
            if (!letterGrade.equals("A")) {
                double nextGradeThreshold = 90;
                if (letterGrade.equals("B")) nextGradeThreshold = 90;
                else if (letterGrade.equals("C")) nextGradeThreshold = 80;
                else if (letterGrade.equals("D")) nextGradeThreshold = 70;
                else nextGradeThreshold = 60;
                
                double pointsNeeded = nextGradeThreshold - average;
                System.out.printf("\nYou need %.2f more points to reach the next grade level.\n", 
                                 pointsNeeded);
            } else {
                System.out.println("\nCongratulations! You have the highest grade!");
            }
            
        } catch (Exception e) {
            System.out.println("\nError: Please enter valid numbers for test scores.");
        } finally {
            scanner.close();
        }
    }
}

// Uncomment to run interactively
InteractiveGradeCalculator.main(null);
System.out.println("Interactive calculator ready! Uncomment the line above to use.");
```
