## Hack 3: Code the Algorithm

Converting the Grade Calculator algorithm into Python code using a function approach:

```java
public class GradeCalculator {
    
    /**
     * Calculate letter grade from three test scores
     * 
     * @param score1 First test score
     * @param score2 Second test score
     * @param score3 Third test score
     * @return Letter grade (String)
     */
    public static String calculateGrade(double score1, double score2, double score3) {
        // Step 1: Add the three scores together
        double total = score1 + score2 + score3;
        
        // Step 2: Calculate the average
        double average = total / 3;
        
        // Step 3: Determine the letter grade using if-else
        String grade;
        if (average >= 90) {
            grade = "A";
        } else if (average >= 80) {
            grade = "B";
        } else if (average >= 70) {
            grade = "C";
        } else if (average >= 60) {
            grade = "D";
        } else {
            grade = "F";
        }
        
        // Step 4: Return the grade
        return grade;
    }
    
    public static void main(String[] args) {
        // Test the function with the provided test cases
        System.out.println("Grade Calculator Tests");
        System.out.println("=".repeat(40));
        System.out.println("Test 1: " + calculateGrade(95, 92, 88));  // Should be 'A'
        System.out.println("Test 2: " + calculateGrade(85, 80, 82));  // Should be 'B'
        System.out.println("Test 3: " + calculateGrade(75, 70, 72));  // Should be 'C'
        System.out.println("Test 4: " + calculateGrade(65, 60, 62));  // Should be 'D'
        System.out.println("Test 5: " + calculateGrade(55, 50, 52));  // Should be 'F'
    }
}

GradeCalculator.main(null);
```

### Enhanced Version with Detailed Output

Here's an improved version that shows the average score along with the letter grade:

```java
public class GradeCalculatorDetailed {
    
    /**
     * Calculate letter grade from three test scores with detailed output
     * 
     * @param score1 First test score
     * @param score2 Second test score
     * @param score3 Third test score
     */
    public static void calculateGradeDetailed(double score1, double score2, double score3) {
        // Calculate total and average
        double total = score1 + score2 + score3;
        double average = total / 3;
        
        // Determine letter grade
        String grade;
        if (average >= 90) {
            grade = "A";
        } else if (average >= 80) {
            grade = "B";
        } else if (average >= 70) {
            grade = "C";
        } else if (average >= 60) {
            grade = "D";
        } else {
            grade = "F";
        }
        
        // Print detailed output
        System.out.printf("Scores (%.0f, %.0f, %.0f) → Average: %.2f → Grade: %s%n",
                         score1, score2, score3, average, grade);
    }
    
    public static void main(String[] args) {
        // Test with detailed output
        System.out.println("\nDetailed Grade Calculator Results");
        System.out.println("=".repeat(40));
        
        // Test cases
        System.out.print("Test 1: ");
        calculateGradeDetailed(95, 92, 88);
        
        System.out.print("Test 2: ");
        calculateGradeDetailed(85, 80, 82);
        
        System.out.print("Test 3: ");
        calculateGradeDetailed(75, 70, 72);
        
        System.out.print("Test 4: ");
        calculateGradeDetailed(65, 60, 62);
        
        System.out.print("Test 5: ");
        calculateGradeDetailed(55, 50, 52);
    }
}

GradeCalculatorDetailed.main(null);
```

### Interactive Version

Bonus: An interactive version where you can input your own scores!

```java
import java.util.Scanner;

public class InteractiveGradeCalculator {
    
    public static void interactiveGradeCalculator() {
        Scanner scanner = new Scanner(System.in);
        
        System.out.println("\n" + "=".repeat(50));
        System.out.println("     INTERACTIVE GRADE CALCULATOR");
        System.out.println("=".repeat(50));
        
        try {
            // Get scores from user
            System.out.print("\nEnter your first test score (0-100): ");
            double score1 = scanner.nextDouble();
            
            System.out.print("Enter your second test score (0-100): ");
            double score2 = scanner.nextDouble();
            
            System.out.print("Enter your third test score (0-100): ");
            double score3 = scanner.nextDouble();
            
            // Validate scores
            if (score1 < 0 || score1 > 100 || score2 < 0 || score2 > 100 || 
                score3 < 0 || score3 > 100) {
                System.out.println("\n⚠️  Error: All scores must be between 0 and 100");
                return;
            }
            
            // Calculate grade
            double total = score1 + score2 + score3;
            double average = total / 3;
            
            String grade;
            if (average >= 90) {
                grade = "A";
            } else if (average >= 80) {
                grade = "B";
            } else if (average >= 70) {
                grade = "C";
            } else if (average >= 60) {
                grade = "D";
            } else {
                grade = "F";
            }
            
            // Display results
            System.out.println("\n" + "-".repeat(50));
            System.out.println("RESULTS:");
            System.out.println("-".repeat(50));
            System.out.printf("Test Scores: %.0f, %.0f, %.0f%n", score1, score2, score3);
            System.out.printf("Average Score: %.2f%n", average);
            System.out.println("Letter Grade: " + grade);
            System.out.println("-".repeat(50));
            
            // Provide feedback
            switch (grade) {
                case "A":
                    System.out.println("🌟 Excellent work! Keep it up!");
                    break;
                case "B":
                    System.out.println("👍 Great job! You're doing well!");
                    break;
                case "C":
                    System.out.println("👌 Good effort! There's room for improvement.");
                    break;
                case "D":
                    System.out.println("⚠️  You passed, but consider studying more for next time.");
                    break;
                case "F":
                    System.out.println("❌ Keep working hard! Consider getting extra help.");
                    break;
            }
            
        } catch (Exception e) {
            System.out.println("\n⚠️  Error: Please enter valid numbers");
        } finally {
            scanner.close();
        }
    }
    
    public static void main(String[] args) {
        // Uncomment the line below to run the interactive calculator
        interactiveGradeCalculator();
        System.out.println("Interactive calculator ready!");
        System.out.println("Uncomment interactiveGradeCalculator() to run.");
    }
}

InteractiveGradeCalculator.main(null);
```

## Summary

### What I Learned:

1. **Algorithms are step-by-step instructions** - Like a recipe, each step must be clear and in the right order
2. **Order matters** - The sequence of steps is crucial for algorithms to work correctly
3. **Functions make code reusable** - Instead of writing the same code multiple times, we can create a function and call it whenever needed
4. **Validation is important** - Checking user input helps prevent errors and makes programs more robust

### Key Takeaways:
- Breaking down complex tasks into simple steps makes them easier to understand and execute
- Clear and specific instructions are essential in both real-world algorithms and programming
- Testing code with multiple cases ensures it works correctly in different scenarios
