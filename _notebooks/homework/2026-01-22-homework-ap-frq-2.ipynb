## Challenge #1

````java
class Scoreboard {

    private String t1Name;
    private String t2Name;
    private int score1;
    private int score2;
    private boolean firstTeamActive;

    public Scoreboard(String t1Name, String t2Name) {
        this.t1Name = t1Name;
        this.t2Name = t2Name;
        this.score1 = 0;
        this.score2 = 0;
        this.firstTeamActive = true;
    }

    public void recordPlay(int p) {
        if (p == 0) {
            firstTeamActive = !firstTeamActive;
        } else {
            if (firstTeamActive) {
                score1 += p;
            } else {
                score2 += p;
            }
        }
    }

    public String getScore() {
        return score1 + "-" + score2 + "-" + (firstTeamActive ? t1Name : t2Name);
    }
}

// Testing the Scoreboard class (DO NOT MODIFY this part unless you change the class, method, or constructer names)
// DO NOT MODIFY BELOW THIS LINE
class Main {
    public static void main(String[] args) {
        String info;

        // Step 1: Create a new Scoreboard for "Red" vs "Blue"
        Scoreboard game = new Scoreboard("Red", "Blue");

        // Step 2
        info = game.getScore();                  // "0-0-Red"
        System.out.println("(Step 2) info = " + info);

        // Step 3
        game.recordPlay(1);

        // Step 4
        info = game.getScore();                  // "1-0-Red"
        System.out.println("(Step 4) info = " + info);

        // Step 5
        game.recordPlay(0);

        // Step 6
        info = game.getScore();                  // "1-0-Blue"
        System.out.println("(Step 6) info = " + info);

        // Step 7 (repeated call to show no change)
        info = game.getScore();                  // still "1-0-Blue"
        System.out.println("(Step 7) info = " + info);

        // Step 8
        game.recordPlay(3);

        // Step 9
        info = game.getScore();                  // "1-3-Blue"
        System.out.println("(Step 9) info = " + info);

        // Step 10
        game.recordPlay(1);

        // Step 11
        game.recordPlay(0);

        // Step 12
        info = game.getScore();                  // "1-4-Red"
        System.out.println("(Step 12) info = " + info);

        // Step 13
        game.recordPlay(0);

        // Step 14
        game.recordPlay(4);

        // Step 15
        game.recordPlay(0);

        // Step 16
        info = game.getScore();                  // "1-8-Red"
        System.out.println("(Step 16) info = " + info);

        // Step 17: Create an independent Scoreboard
        Scoreboard match = new Scoreboard("Lions", "Tigers");

        // Step 18
        info = match.getScore();                 // "0-0-Lions"
        System.out.println("(Step 18) match info = " + info);

        // Step 19: Verify the original game is unchanged
        info = game.getScore();                  // "1-8-Red"
        System.out.println("(Step 19) game info = " + info);
    }
}

Main.main(null);
````

## How I Solved it

I began by identifying the necessary instance variables. The prompt specified two team names, so I created `t1Name` and `t2Name`. I also needed to track scores, initializing `score1` and `score2` to zero. To manage the turns, I used a boolean `firstTeamActive`, setting it to true initially.

For the `recordPlay` method, I checked if the points argument `p` was zero. If so, I toggled the `firstTeamActive` flag. If points were scored, I added them to the currently active team's score using the boolean to direct the addition to the correct variable.

Finally, for `getScore`, I constructed the required string by concatenating the scores and hyphens. I used the ternary operator to conciseley determine which team name to append based on the `firstTeamActive` state.

## Area of Struggle

My main issue was in the `getScore` method. I originally tried to write an if-else block to return the string, which made the code very clunky. I also briefly struggled with the output format; I forgot that the active team's name needs to be at the end, and I was printing the boolean value `true` or `false` by mistake. 

My initial incorrect return statement looked something like this:

```java
// Incorrect attempt
return score1 + "-" + score2 + "-" + firstTeamActive;
```

This outputted results like `0-0-true` instead of `0-0-Red`.

I fixed this by deciding to use the ternary operator `(firstTeamActive ? t1Name : t2Name)` to cleanly select the right name inline.
